```python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
from keras.callbacks import LambdaCallback
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
X,y = make_classification(n_samples=500)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.33,random_state=12238)
```

```python
sgd = SGD()

model = Sequential()
model.add(Dense(20,input_dim=20,activation = 'softmax'))
model.compile(loss= 'sparse_categorical_crossentropy',optimizer= sgd)
```

```python
weight = []
print_weight = LambdaCallback(on_epoch_end=lambda epoch,logs:weight.append(model.layers[0].get_weights()))
```

```python
model.fit(X_train,y_train,batch_size=X_train.shape[0],epochs=100,verbose=0,validation_data=(X_test,y_test),
          callbacks=[print_weight])
```

```python
w_all_data = [item[0] for item in [item[0] for item in [item[0] for item in weight]]]
```

```python
plt.plot(w_all_data)
```
