##### Entrenar un modelo que sea capas de predecir

```python
#Se eliminan estas columnas, ya que estas nos existen en el df_test
df = df.drop(columns=['offshld_Y', 'addrpct', 'asltweap_Y', 'stinter_other', 'crossst_other', 'trhsloc_U','city_STATEN ISLAND', 'crimsusp_other', 'premname_other', 'linecm', 'officrid_Y', 'offverb_Y', 'detailcm', 'haircolr_SN'])
```

```python
X = df.drop(columns='arstmade')
Y = df['arstmade']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=.33,random_state=11238)
```

```python
parameter_Gradient={'learning_rate': [0.3, 0.5,0.8],'n_estimators': [500, 1000, 1500],'subsample': [0.3,0.5,0.6]}
cl_Gradient = GridSearchCV(estimator=GradientBoostingClassifier(), param_grid=parameter_Gradient, n_jobs=-1,cv=2)
cl_Gradient.fit(X_train, y_train)
model = cl_Gradient.best_estimator_
```

```python
model = cl_Gradient.best_estimator_
```

##### Para ver el rendimiento del modelo se utiliza clasification report, el que entrega metricas como accuracy y a la vez una matriz de confucion para ver la cantidad de respuestas correctas

```python
y_hat = model.predict(X_test)
print(classification_report(y_test,y_hat))
```

```python
confusion_matrix(y_test,y_hat)
```

Se observa este comportamiento devido a que la mayor parte de las veces apunta a que no fue arrestado y como existe un desvalanceo de la clase tiende a acertar

###### Se mostraran los atributos mas significativos y luego se entregara el nombre de los 30 mayores

```python
def plot_feature_importance(fit_model, feat_names):
    tmp_importance = fit_model.feature_importances_
    sort_importance = np.argsort(tmp_importance)[::-1]
    names = [feat_names[i] for i in sort_importance]
    plt.barh(
        range(len(feat_names)), tmp_importance[sort_importance]
    )
    plt.yticks(range(len(feat_names)),
               names, rotation=0)
    
    return names
```

```python
plt.figure(figsize=(30,30))
thirty = plot_feature_importance(model,X.columns)
```

###### Los 30 mayores corresponden a (de mayor a menor):

```python
thirty = thirty[:30]
thirty
```
