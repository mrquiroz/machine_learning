```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.simplefilter('ignore')
plt.rcParams['figure.figsize']=(10,6)
plt.style.use('seaborn')
```

Se trajara con la base de datos iris

```python
df = pd.read_csv('iris.csv').drop(columns='Unnamed: 0')
```

```python
df.sample()
```

```python
for n,i in enumerate(df.drop(columns='Species').columns):
    plt.subplot(2,2,n+1)
    df.groupby('Species')[i].plot(kind='hist',alpha=.5)
    plt.legend(df['Species'].unique())
    plt.title(i)
    plt.tight_layout()
```

Se observa que para el sepalo existe una fuerte superpocicion de clases, lo que puede significar que discriminar bajo esta caracteristica puede resultar dificil, por su parte el petalo es mas informativo y se observa que la especie "setosa" es mas diferente

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
```

```python
df.Species.value_counts()
```

```python
df['Species'] = LabelEncoder().fit_transform(df['Species'])
```

```python
df.Species.value_counts()
```

```python
X_train, X_test, y_train,y_test = train_test_split(df[df.columns[:-1]],df['Species'],test_size=.33,random_state=11238)
```

```python
lda_model = LinearDiscriminantAnalysis()
lda_model.fit(X_train,y_train)
```

```python
lda_model.priors_.round(2)
```

```python
lda_model.means_.round(2)
```

La 1ra fila corresponde a setosa, se observa que sus valores difieren un poco de los otros 2

```python
lda_class_pred = lda_model.predict(X_test)
```

```python
lda_class_pred
```

```python
import lec4_graphs as afx
```

```python
targets = ['setosa','versicolor','virginica']
targets
```

```python
afx.plot_confusion_matrix(y_test,lda_class_pred,targets)
```

```python
afx.plot_class_report(y_test,lda_class_pred,targets)
```
